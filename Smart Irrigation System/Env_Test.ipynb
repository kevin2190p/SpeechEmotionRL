<h2><span style="font-family:Winky Rough">2. Environment Testing</span></h2>

```python
env_arid = SimpleCornIrrigationEnv(region_type="arid", render_mode="human")
env_temperate = SimpleCornIrrigationEnv(region_type="temperate", render_mode="human")
env_tropical = SimpleCornIrrigationEnv(region_type="tropical", render_mode="human")
```

```python
#  Using a randomized action test environment
def test_environment(env, n_steps=10):
    observation,_ = env.reset()
    print(f"Initial observation {observation}")
    
    total_reward = 0
    for step in range(n_steps):
        action = np.random.randint(0, 4)
        observation, reward, done, _, info = env.step(action)
        total_reward += reward
        env.render()
        print(f"Step {step}, Action: {action}, Reward: {reward: .2f}")
        
        if done:
            print("Episode finished early!")
            break
    
    print(f"Total Reward: {total_reward:.2f}")
    env.plot_episode()
        
```

```python
for env in [env_arid, env_temperate, env_tropical]:
    print(f"\nTesting {env.region_type} environment...")
    test_environment(env, 120)
```
